# Notebook de analisis de datos

Este notebook tiene como objetivo dar un paso a paso en el proceso del analisis de datos, para su observación, limpieza y procesamiento para ser usados en algoritmos de Machine Learning.

Usa el dataset de precios de casa de Ames para técnicas de regresiones avanzadas.  

## Descripción general

Este notebook muestra y trata de explicar el proceso de analisis de datos, su objetivo es el encapsular cada parte del proceso para ser usado en cualquier problema en general.

El proceso es el siguiente:

* Imputación de valores omitidos de manera sequencial en la data. 

* Transformación de algunos valores numericos que en realidad son categoricos. 

* Codificación de etiquetas de algunas variables categoricas que puedan contener información jerarquica en su set

* Transformación de caracterisiticas asimetricas usando Box Cox Transformation, con mejores resultados que la logaritmica

* Obtención de variables categoricas con dummy features.

El objetivo de este notebook es hacer más entendible las técnicas de limpieza de datos, por eso mismo aunque las técnicas son las mismas el orden y estructura del notebook difiere de lo encontrado en los scripts.

## Entendiendo el problema

Este es un dataset extenso cuenta con muchas caracteristicas, esto lo hace bastante complicado. En el ML y la ciencia de datos, la información base es lo más importante. Una recomendación general para todos los problemas es entender la naturaleza de los mismos.

Para esto se recomienda observar cada variable y entender su significado y relevancia en el problema. A mí me funciono definir en una hoja de excel cada variable y relacionarlas de la siguiente manera:

* Variable: Nombre de la variable

* Tipo: Identificación del tipo de variable. Existen dos tipos categoricas y numericas

* Segmento: Se puede definir tres tipos de segmentos caracteristicas fisicas, caracteristicas de espejo, caracteristicas de localización.

* Expectativa: La expectativa que se tiene sobre la influencia de una variable independiente sobre la independiente: Alta, Media, Baja.

* Conclusión: Las conclusiones obtenidas una vez se analisa la variable, se llena de la misma forma que expectativa.

Mientras que tipo y segmento son valores para futuras referencias, la columna expectativa es importante para entender la relación de las caracteristicas con el problema, para llenarlo sirve mucho preguntarse:

* ¿Podria esta variable incidir en el resultado final de la variable.

* Si es así ¿Qué tanto lo haría?

* ¿Está la información de esta variable descrita en otra?

## Analizando la variable objetivo

Es muy importante entender como se comporta estadisticamente la variable objetivo para problemas de regresión y clasificación.

Para los problemas de clasificación es recomendable escalar la variable entre rangos pequeños o similares a los de todo el dataset para evitar resultados pobres. Las escalas comunes son [-1, 1] o [-3, 3]. Para esto se usa el metodo StandardScaler de Scikit-learn.

Para los problemas de regresión se busca normalizar la variable para que esta tenga un comportamiento linear.

Este notebook se enfoca en regressión, sin embargo la función de StandardScaler está en los srcripts.

```python
sns.distplot(train['SalePrice'] , fit=norm);

# Obtener los parametros mu y sigma usados en la función 
(mu, sigma) = norm.fit(train['SalePrice'])
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

#Dibujo de la distribución normal, contra la variable
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
plt.ylabel('Frequency')
plt.title('SalePrice distribution')

#Plot de probabilidad
fig = plt.figure()
res = stats.probplot(train['SalePrice'], plot=plt)
plt.show()
```

De estas graficas se observa que la variable objetivo tiene una asimetria hacia la derecha y dado que los modelos de regresión trabajan mejor con valore normalizados, se hace nesesario normalizar la variable. Para esto se aplica la transformación logaritmica.

```python
#Se usa la función numpy log1p que aplica log(1+x) a todos los elementos de la columna
train["SalePrice"] = np.log1p(train["SalePrice"])

#Se verifica la nueva distribución 
sns.distplot(train['SalePrice'] , fit=norm);

# Obtener los parametros mu y sigma usados en la función
(mu, sigma) = norm.fit(train['SalePrice'])
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

#Dibujo de la distribución normal, contra la variable
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
plt.ylabel('Frequency')
plt.title('SalePrice distribution')

#Plot de probabilidad
fig = plt.figure()
res = stats.probplot(train['SalePrice'], plot=plt)
plt.show()
```

## Imputando variables faltantes

Existen dos preguntas importantes sobre la limpieza de la data faltante:

* ¿Qué tan importante es la data faltante?

* La data faltante ¿Es aleatoria o tiene un patrón?

La respuesta a esta preguntas es importante por razones practicas, dado que data faltante puede implicar una reducción del tamaño de la muestra que impactara el analisís final. También desde una perspectiva substancial es nesesario asegurar que la data faltante no genera cesgo o esconde una caracterisitca importante del dataset.

#### Pool QC

La descripción del dataset dice que el valor NaN se refiere a sin piscina, en ese orden de ideas mejor crear un valor categorico de este tipo que eliminar la variable.

```python
train["PoolQC"] = train["PoolQC"].fillna("None")
test["PoolQC"] = test["PoolQC"].fillna("None")
```

#### Alley

#### Fireplace Qu

```python
train["FireplaceQu"] = train["FireplaceQu"].fillna("None")
test["FireplaceQu"] = test["FireplaceQu"].fillna("None")
```

#### LotFrontage 

Dado que el area de cada calle está conectado al de la propiedad de la casa, es muy posible que comparta esta caracterisitca con otras casa del mismo barrio. Bajo este escenario se puede imputar este valor con la media de las areas agrupadas por barrio, apoyandose en la caracteristica barrio.

```python
#Agrupar por barrios e imputar los valores faltantes por la media de LotFrontage de todo el barrio.
train["LotFrontage"] = train.groupby("Neighborhood")["LotFrontage"].transform(
    lambda x: x.fillna(x.median()))
test["LotFrontage"] = test.groupby("Neighborhood")["LotFrontage"].transform(
    lambda x: x.fillna(x.median()))
```

### Creando caracteristicas artificiales

Esto se usa mucho para manejar variables categoricas, pero también se puede usar para crear variables artificiales, convertir variables numericas a categoricas por su comportamiento o definición (fechas) y manejar caracterisitcas asimetricas, esto ultimo solo aplica para regresión. Para clasificación se escala con StandardScaler.

#### Transformación de variables numericas a categoricas

```python
#MSSubClass=La clase de construcción
train['MSSubClass'] = train['MSSubClass'].apply(str)
test['MSSubClass'] = test['MSSubClass'].apply(str)


#Cambiando OverallCond a una variable categorica
train['OverallCond'] = train['OverallCond'].astype(str)
test['OverallCond'] = test['OverallCond'].astype(str)


#Caracteristicas de tiempo.
train['YrSold'] = train['YrSold'].astype(str)
train['MoSold'] = train['MoSold'].astype(str)

test['YrSold'] = test['YrSold'].astype(str)
test['MoSold'] = test['MoSold'].astype(str)
```

#### Transformación con label encoding para variables categoricas, solo aplica cuando contiene información que las puede categorizar en clases

```python
from sklearn.preprocessing import LabelEncoder
cols = ('FireplaceQu', 'BsmtQual', 'BsmtCond', 'GarageQual', 'GarageCond', 
        'ExterQual', 'ExterCond','HeatingQC', 'PoolQC', 'KitchenQual', 'BsmtFinType1', 
        'BsmtFinType2', 'Functional', 'Fence', 'BsmtExposure', 'GarageFinish', 'LandSlope',
        'LotShape', 'PavedDrive', 'Street', 'Alley', 'CentralAir', 'MSSubClass', 'OverallCond', 
        'YrSold', 'MoSold')

for c in cols:
    lbl = LabelEncoder() 
    lbl.fit(list(train[c].values)) 
    train[c] = lbl.transform(list(train[c].values))
    
    lbl.fit(list(test[c].values)) 
    test[c] = lbl.transform(list(test[c].values))
```

#### Agregando caracteristicas nuevas

```python
train['TotalSF'] = train['TotalBsmtSF'] + train['1stFlrSF'] + train['2ndFlrSF']
test['TotalSF'] = test['TotalBsmtSF'] + test['1stFlrSF'] + test['2ndFlrSF']
```

#### Caracteristicas asimetricas

```python
numeric_feats = train.dtypes[train.dtypes != "object"].index

skewed_feats = train[numeric_feats].apply(lambda x: skew(x.dropna())).sort_values(ascending=False)
print("\nSkew in numerical features: \n")
skewness_train = pd.DataFrame({'Skew' :skewed_feats})
print(skewness_train.head(10))

numeric_feats = test.dtypes[test.dtypes != "object"].index

skewed_feats = test[numeric_feats].apply(lambda x: skew(x.dropna())).sort_values(ascending=False)
print("\nSkew in numerical features: \n")
skewness_test = pd.DataFrame({'Skew' :skewed_feats})
print(skewness_test.head(10))
```

#### Transformación BoxCox para categorias con alta asimetria, aplica solo para regressión. Para clasificación es mejor escalar

En este paso se usa la función scipy que computa la Box-Cox transformada de 1 + x. Más información en esta pagina:

http://onlinestatbook.com/2/transformations/box-cox.html

```python
skewness_train = skewness_train[abs(skewness_train) > 0.75]
print("There are {} skewed numerical features to Box Cox transform".format(skewness_train.shape[0]))

from scipy.special import boxcox1p
skewed_features = skewness_train.index
lam = 0.15
for feat in skewed_features:
    train[feat] = boxcox1p(train[feat], lam)

skewness_test = skewness_test[abs(skewness_test) > 0.75]
print("There are {} skewed numerical features to Box Cox transform".format(skewness_test.shape[0]))

from scipy.special import boxcox1p
skewed_features = skewness_train.index
lam = 0.15
for feat in skewed_features:
    test[feat] = boxcox1p(test[feat], lam)
```

#### Obteniendo categorias dummy

```python
train = pd.get_dummies(train)
print(train.shape)

test = pd.get_dummies(test)
print(test.shape)
```

#### Filtro de las caracteristicas dummy de train que no coinciden con las de test.

```python
drop_col = [list(train.columns)[i] for i in range(len(list(train.columns))-1) if list(train.columns)[i] not in test.columns] 
train.drop(drop_col, axis=1, inplace=True)
```
